# Génération d'un fichier csv
### L'objectif de ce script est de charger les données en ligne de OpenFoodFacts pour les transformer en un énorme fichier csv.

```python
import pandas as pd
import requests
from io import BytesIO
import gzip
```

## Panneau de configuration

```python
project_path = "C:\\Users\\charl\\Documents\\workspace\\green_ia\\cch\\"
file_nbr = '01' # numéro d'identification des csv à générer 
```

```python
# récupére données open food facts et en génére un df en local (inutile de l'exex systématiquement...)
url = "https://static.openfoodfacts.org/data/en.openfoodfacts.org.products.csv.gz"
response = requests.get(url)

if response.status_code == 200:
    with gzip.GzipFile(fileobj=BytesIO(response.content), mode='rb') as file:
        df = pd.read_csv(file, sep='\t', encoding='utf-8')
        pd.set_option('display.max_columns', None)
        print(df)
        
        df.to_csv(project_path + f'data_global\\openfoodfacts_{file_nbr}.csv', index=False)  # Vous pouvez spécifier l'index comme False si vous ne voulez pas le sauvegarder
        print("csv sauvegardé en local")
else:
    print(f"error, {response.status_code}.")
```
